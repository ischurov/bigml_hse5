```python
P = np.array([2, 3])
T = np.array([5, 4])
u = T - P
ts = np.linspace(-5, 5)
```

We are considering line $\{P + t\vec{u}\mid t \in \mathbb R\}$.

```python
Qs = []
for t in ts:
    Qs.append(P + t * u)
```

```python
Qs = [P + t * u for t in ts]
```

```python
plt.plot([q[0] for q in Qs], [q[1] for q in Qs])
plt.plot(P[0], P[1], 'o')
plt.plot(T[0], T[1], 'o')
```
